```python
import numpy as np
import pandas as pd
import csv
```

# processing data

```python
def DataProcess(Numatoms, list):
    Num = sum(Numatoms)
    Len = len(Numatoms)
    Name = list.columns[1].split('_')[-1] 
    ListAverge = list.astype(float).mean(axis=1)    # compute the average
    ListAvergeW = np.array(list.astype(float)) * np.array(Numatoms)  # compute the weighted average 
    ListAvergeW = np.sum(ListAvergeW, axis=1) / Num
    ListMax = list.astype(float).max(axis=1)    # compute the maximum 
    ListMin = list.astype(float).min(axis=1)    # compute the minimum
    ListStd0 = list.astype(float).std(axis=1, ddof=0)  # compute biased standard Deviation (/n)
    ListStd1 = list.astype(float).std(axis=1, ddof=1)  # compute unbiased standard Deviation (/(n-1))
    ListStd0W = ((np.array(list.astype(float)) - np.array([ListAvergeW] * 3).T) ** 2) / Len    # compute weighted and biased standard Deviation
    ListStd0W = np.sqrt(np.sum(ListStd0W, axis=1))
    ListStd1W = ((np.array(list.astype(float)) - np.array([ListAvergeW] * 3).T) ** 2) / (Len - 1)    # compute weighted and biased standard Deviation
    ListStd1W = np.sqrt(np.sum(ListStd1W, axis=1))
    ListAverge = pd.DataFrame(ListAverge, columns=['Aver_' + Name])
    ListAvergeW = pd.DataFrame(ListAvergeW, columns=['AverW_' + Name],
                               index=[i + 1 for i in range(list.shape[0])])
    ListMax = pd.DataFrame(ListMax, columns=['Max_' + Name])
    ListMin = pd.DataFrame(ListMin, columns=['Min_' + Name])
    ListStd0 = pd.DataFrame(ListStd0, columns=['Std0_' + Name])
    ListStd1 = pd.DataFrame(ListStd1, columns=['Std1_' + Name])
    ListStd0W = pd.DataFrame(ListStd0W, columns=['Std0W_' + Name],
                             index=[i + 1 for i in range(list.shape[0])])
    ListStd1W = pd.DataFrame(ListStd1W, columns=['Std1W_' + Name],
                             index=[i + 1 for i in range(list.shape[0])])
    ListFinal = pd.concat([list, ListAverge, ListAvergeW, ListMax, ListMin, ListStd0,
                           ListStd1, ListStd0W, ListStd1W], axis=1)
    ListFinal = ListFinal.iloc[:, 3:]
    return ListFinal
```

# read sample data

```python
ElementList = pd.read_csv('Sample with GH.csv',header = None)
ElementList.columns = ElementList.loc[0, :]
ElementList = ElementList.drop([0])
ElementList
```

# read element data

```python
DescriptorList = pd.read_csv('Periodic table information.csv',header = None)
DescriptorList.columns = DescriptorList.loc[0, :]
DescriptorList = DescriptorList.drop([0])
DescriptorList
```

# split MAZ name

```python
SampleList = ElementList['Sample'].str.split('_', expand=True)
NumElement = SampleList.shape[-1]
NumFeature = DescriptorList.shape[-1] - 1
NumSample = SampleList.shape[0]
```

# generate primary features of MAZ

```python
List = ElementList.copy()
for i in range(NumElement):
    for j in DescriptorList.columns[1:]:
        List[str(i + 1) + '_' + j] = ''
        for k in range(NumSample):
            List.loc[k + 1,str(i + 1) + '_' + j] = str(float(DescriptorList[DescriptorList['Element'].isin([SampleList[i][k + 1].strip()])][j]))
List
```

# generate primary features using statistical functions of MAZ

```python
AtomNumList = [1, 2, 4]
Data = List
for i in range(NumFeature):
    DataSep = []
    for j in range(NumElement):
        DataSep.append(List[str(j + 1) + '_' + List.columns[i + 2].split('_')[-1]])
    DataSep = pd.DataFrame(DataSep).T
    DataSep = DataProcess(AtomNumList, DataSep)
    Data = pd.concat([Data, DataSep], axis=1)
Data
```

# output features

```python
List.to_csv("Primary features.csv", index=False, sep=',')
Data.to_csv("Primary features using statistical functions.csv", index=False, sep=',')
```
